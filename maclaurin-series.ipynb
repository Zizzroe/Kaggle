```python
import numpy as np
import matplotlib.pyplot as plt
import math

def factorial(num):
    return math.factorial(num)

def maclaurin(bound, x):
    if bound < 1:
        return 0
    approx = 0
    limit = bound
    for n in range(limit):
        approx += (pow(x, n)) / factorial(n)
    return approx

for n in range(1, 10):
    def maclaurinInterval():
        values = []
        # Reduce n to avoid OverflowError
        for x in np.arange(-4, 2, 0.1):
            values.append((x, maclaurin(n, x)))
        return values

    def actual_exponential(x):
        return np.exp(x)

    plot_data = maclaurinInterval() 

    x_values, y_values = zip(*plot_data)
    actual_values = actual_exponential(np.array(x_values))

    plt.plot(x_values, y_values, label='Maclaurin Series')
    plt.plot(x_values, actual_values, label='Actual e^x', linestyle='--')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Maclaurin Series vs Actual e^x')
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#get_user_input = int(input("Number of Iterations: "))
 
for get_user_input in range(1, 10):    
    def lnTaylor(bound, x):
        if bound < 1:
            return 0
        limit = bound
        approx = 0
        for n in range(1, limit):
            approx += pow(-1, n + 1) * pow(x, n) / n
        return approx

    def maclaurinInterval():
        values = []
        n = get_user_input  # Reduce n to avoid OverflowError
        for x in np.arange(1 * pow(10, -9), 2, 0.001): #decrese gap to infinity by decresing the power
            values.append((x, lnTaylor(n, x)))
        return values

    def actualln(x):
        return np.log(x)

    def asymptote():
        values_asymptote = []
        for n in np.arange(-20, 40, 0.01):
            values_asymptote.append((1, n))
        return values_asymptote

    plot_data = maclaurinInterval() 
    asymptote_data = asymptote()

    x_values_asymptote, y_values_asymptote = zip(*asymptote_data)
    x_values, y_values = zip(*plot_data)
    actual_values = actualln(np.array(x_values))

    plt.plot(x_values, y_values, label='Maclaurin Series')
    plt.plot(x_values, actual_values, label='Actual lnx', linestyle='--')
    plt.plot(x_values_asymptote, y_values_asymptote, label='x → 1+, % error → ∞', linestyle='-.')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Maclaurin Series vs Actual lnx')
    plt.legend()
    plt.show()
```
